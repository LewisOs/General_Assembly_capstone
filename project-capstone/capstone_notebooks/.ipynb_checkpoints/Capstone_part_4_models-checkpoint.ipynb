## Feature engineering

#### Polynomial features

As discussed during the EDA, all of the categorical variables have non-linear relationships with the target variable. As such, I want to include polynomial features for these.

```python
# adding a litter size feature 
pets_df['litter_size'] = pets_df['males_in_litter'] + pets_df['females_in_litter']
```

```python
# creating polynomial features
from sklearn.preprocessing import PolynomialFeatures

# variables I want polynomials for
cont_vars = ['litter_size', 
             'n_images', 
             'seller_n_adverts',
             'pets_age_in_days', 
             'males_in_litter',
             'females_in_litter']

# using 2nd order polynomials, based on EDA
poly = PolynomialFeatures(degree = 2, include_bias = False)

# fitting transformer
polyfeats = pd.DataFrame(poly.fit_transform(pets_df[cont_vars]))

# I only want the polynomials, no interactions
poly_indexes = [i for i, e in enumerate(poly.get_feature_names_out()) if '^' in e]
polyfeats = polyfeats[poly_indexes]

# creating a dict to update the column names
ployfeats_names = {}
for i, n in zip(poly_indexes, cont_vars):
    ployfeats_names[i] = n + '_poly'

# updating column names
polyfeats.rename(columns = ployfeats_names, inplace = True)

# combinging with dataset
pets_df = pets_df.join(polyfeats)
```

#### Dummification

```python
pets_df = pd.get_dummies(pets_df, drop_first = True)
pets_df.reset_index(inplace = True)
pets_df.drop(columns = 'index', inplace = True)
```

## Splitting and scaling the data

```python
# splitting the predictors and target
X = pets_df.drop(columns = 'price')
y = pets_df['price']
```

```python
# splitting into train and test sets
from sklearn.model_selection import train_test_split

# creating train & test sets. n.b. data is roughly frequency binned by y
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = .2, 
                                                    random_state = 3,
                                                    stratify = pd.qcut(y,50, duplicates ='drop')
                                                   )
    
```

```python
# value counts of categories post data split - sense check
# also means of target in train and test
pd.DataFrame({'y':y.describe(), 'y_train':y_train.describe(), 'y_test':y_test.describe()})
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Building models

### Initial models

I'm going to begin by trying a large number of model types with some fairly standard values to see what kinds of baseline performance I can expect from each model type. After doing this, I will select the 3-5 most promising models and work on parameter tuning. I will then perfom a deeper analysis on the best of the tuned models. Specifically, this will involve looking at the coefficients/feature importances to determine which are most influential on the model's predictions and an analysis of the residuals. Finally, I will build a stacking ensemble model from the 3-5 best tuned models to see whether this can out perform the single best model.

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import RidgeCV, LassoCV, ElasticNetCV
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor, AdaBoostRegressor, GradientBoostingRegressor
```

```python
models = [RidgeCV(cv = 5),
         LassoCV(cv = 5, n_jobs=8),
         ElasticNetCV(n_alphas=100, cv=5,verbose=1, n_jobs=8),
         KNeighborsRegressor(),
         DecisionTreeRegressor(max_depth = 10),
         RandomForestRegressor(max_depth = 10),
         ExtraTreesRegressor(max_depth = 10), 
         AdaBoostRegressor(DecisionTreeRegressor(max_depth=10), n_estimators = 200, learning_rate = 0.5), 
         GradientBoostingRegressor(max_depth = 10, learning_rate = 0.25)]

res = []
for model in models:
    print(model)
    print()
    model.fit(X_train, y_train)
    train_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)
    print('Train score:', train_score)
    
    print('Test score:', test_score)
    res.append([model, train_score, test_score])
    print()
    print('-'*60)
    print()
```

```python
for model in res:
    print(model)
```

#### Elastic Net

##### Parameter tuning for elastic net regression

As with the initial LassoCV model, the initial ElasticNetCV model would have done some basic parameter tuning over the values of alpha. So, similarly to above, rather than using grid search, I will use the ElasticNetCV class with a larger n_alphas.

ElasticNetCV also allows parameter tuning over the values of l1_ratio (i.e. the ration of ridge and lasso regression). In the intial model, this would have been the default value (0.5). The documentation for the class notes that 'a good choice of list of values for l1_ratio is often to put more values close to 1 (i.e. Lasso) and less close to 0 (i.e. Ridge)', and I have chosen the parameter range accordingly.

```python
# defining the model 
elastic = ElasticNetCV(cv = 3, 
                         n_alphas = 1000, 
                         max_iter = 100000, 
                         n_jobs=-1, 
                         l1_ratio = [.1, .5, .7, .9, .95, .99])

# fitting the model
elastic.fit(X_train, y_train)

# evaluating the model
print(elastic.score(X_train, y_train))
print(elastic.score(X_test, y_test))
print(elastic.alpha_)
print(elastic.l1_ratio_)
```

Tuning the alpha value for the elastic net model has resulsed in a very small improvement on the test set (i.e. from 0.62 to 0.63). Interestingly, the best alpha level is quite a bit smaller than that for the LASSO regrsssor. The best l1_ratio value (l1_ration) is also greater than the default value. This means that the model is using more l1 than l2 regularisation (given the relative performances of the the initial ridge and LASSO models, this is to be expected).

I will also see whether this model can be futher improved through the random patches method. I am using a smaller number of estimators than before, due to long training times.

#### Random forest regressor

##### Parameter tuning for random forest regresssor

For the random forest regressor, I am attempting to find a set of parameters which can attain higher test set R2 than the initial model's (0.61) and ideally greater than the inital gradient boost regressor model (0.64). 

I'm going focus the grid search on 3 parameters used for regaularisation; max_depth, min_samples_split and max_leaf_nodes. I'll begin by looking over a reasonably wide range of values to understand the ideal scale for each parameter.

```python
# defining the model 
rand_for = RandomForestRegressor()

# defining param grid
rand_for_params = {
    'max_depth' : [10,20,40],
    'min_samples_split' : [4,8,16],
    'max_leaf_nodes' : [100, 200, 400]
}

# instantiate the grid search 
rand_for_gridsearch = GridSearchCV(rand_for, rand_for_params, n_jobs=8, cv=3, verbose=1, return_train_score=True)

# find the fit in the gridsearch 
rand_for_gridsearch.fit(X_train, y_train)

print(rand_for_gridsearch.score(X_test, y_test))

print('Best estimator:', rand_for_gridsearch.best_estimator_)
print('Mean cross-validated score of the best estimator: ', rand_for_gridsearch.best_score_)
```

The best estimator found by the 1st round of parameter tuning has reached a test set R2 of 0.63. This is a small improvement on the initial model, but still slightly behind the initial gradient boosting regressor. The best estimators parameters are max_depth=40, max_leaf_nodes=400, min_samples_split=16. This means that for each parameter, the highest value searched over was found to be the best. This indicates that it may be possible to attain slightly better performance by searching over a range of larger values.

```python
# defining param grid
rand_for_params = {
    'max_depth' : [40,80],
    'min_samples_split' : [16,32],
    'max_leaf_nodes' : [400, 800, 1600]
}

# instantiate the grid search 
rand_for_gridsearch = GridSearchCV(rand_for, rand_for_params, n_jobs=8, cv=3)

# find the fit in the gridsearch 
rand_for_gridsearch.fit(X_train, y_train)

print(rand_for_gridsearch.score(X_test, y_test))

print('Best estimator:', rand_for_gridsearch.best_estimator_)
print('Mean cross-validated score of the best estimator: ', rand_for_gridsearch.best_score_)
```

```python
from sklearn.ensemble import StackingRegressor

stack_reg = StackingRegressor(estimators = [('lasso', Lasso(alpha = 4.603838322648223)),
                                           ('elastic', ElasticNet(alpha = 3.028253683790783, l1_ratio = 0.95),
                                           'random_forest', RandomForestRegressor(max_depth=80, max_leaf_nodes=1600, min_samples_split=16),
                                           'gbregressor', GradientBoostingRegressor(max_depth = 4, learning_rate = 0.30, n_estimators = 200, min_samples_split = 16))],
                             final_estimator = GradientBoostingRegressor(),
                             cv = None)

stack_reg.fit(X_train, y_train)
print(stack_reg.score(X_test, y_test))
```

I've created a stacking ensemble with the four best models as the base estimators and a gradient boosting regressor as the meta-learner. The model has a worse test set R2 than the gradient boosting regressor. Fruther analysis of the distribution of errors between models may be able to explain why this is the case.
